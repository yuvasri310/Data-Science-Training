CSV OPERATIONS

```python
#1
import pandas as pd
import matplotlib.pyplot as plt
csv_data = """Name,Math,Science,English
Amit,78,85,74
Priya,92,89,96
Kiran,65,70,60
Anita,88,95,90
Ravi,55,60,58
"""
with open('students.csv', 'w') as f:
    f.write(csv_data)
df = pd.read_csv('students.csv')
print("First 3 rows:")
print(df.head(3))
```

```python
#2
df['Average']=df[['Math','Science','English']].mean(axis=1)
top_student=df.loc[df['Average'].idxmax()]
print("Student with highest average:")
print(top_student)
```

```python
#3
print("students score more than 80 in science")
print(df[df['Science']>80])
```

```python
#4
print("English score top 3 students desc")
print(df.sort_values(by='English',ascending=False).head(3))
```

```python
#5
plt.figure(figsize=(8,5))
plt.bar(df['Name'],df['Average'],color='red')
plt.title('Students Average Scores')
plt.xlabel('Name')
plt.ylabel('Average Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.set_index('Name')[['Math', 'Science', 'English']].plot(kind='line', marker='o')
plt.title('Student Scores in Math, Science, and English')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#6
totals = df[['Math', 'Science', 'English']].sum()
top_subject = totals.idxmax()
print("\nSubject with the highest total score:", top_subject)
df.to_csv('students_summary.csv', index=False)
```
