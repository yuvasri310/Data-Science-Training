Monthly Retail Sales Analysis

```python
#1
import pandas as pd
import matplotlib.pyplot as plt
csv_data = """Month,Store,Sales
Jan,Store A,12000
Jan,Store B,15000
Feb,Store A,18000
Feb,Store B,17000
Mar,Store A,16000
Mar,Store B,19000
"""
with open('sales_data.csv', 'w') as file:
    file.write(csv_data)
df = pd.read_csv('sales_data.csv')
print("Full DataFrame:")
print(df)

```

```python
#2
total_sales = df.groupby("Store")["Sales"].sum()
print("\nTotal Sales per Store:")
print(total_sales)
higher_store = total_sales.idxmax()
print(f"\nStore with higher total sales: {higher_store}")
```

```python
#3
monthly_sales = df.groupby("Month")["Sales"].sum()
print("\nMonthly Sales Trend (Total Sales per Month):")
print(monthly_sales)
```

```python
#4
pivot_df = df.pivot(index='Month', columns='Store', values='Sales')
pivot_df.plot(marker='o')
plt.title("Monthly Sales by Store")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.legend(title="Store")
plt.show()
```

```python
#5
total_sales.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Total Sales per Store")
plt.xlabel("Store")
plt.ylabel("Sales")
plt.xticks(rotation=0)
for i, value in enumerate(total_sales):
    plt.text(i, value + 500, str(value), ha='center')
plt.tight_layout()
plt.savefig("sales_chart.png")
plt.show()
```
